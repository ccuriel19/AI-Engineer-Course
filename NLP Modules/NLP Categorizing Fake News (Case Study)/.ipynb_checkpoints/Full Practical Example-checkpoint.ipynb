```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
from spacy import displacy
from spacy import tokenizer
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import gensim
import gensim.corpora as corpora
from gensim.models.coherencemodel import CoherenceModel
from gensim.models import LsiModel, TfidfModel
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# initialize spacy model:
nlp = spacy.load('en_core_web_sm')

fake_news = data[data['fake_or_factual'] == 'Fake News']
fact_news = data[data['fake_or_factual'] == 'Factual News']

# Create two seperate spacy documents with just the text of the articles:
fake_spacydocs = list(nlp.pipe(fake_news['text']))
fact_spacydocs = list(nlp.pipe(fact_news['text']))

# Create function to extract the tags
def extract_token_tags(doc:spacy.tokens.doc.Doc):
    return[(i.text, i.ent_type_, i.pos_) for i in doc]

fake_tagsdf = []
columns = ['token', 'ner_tag', 'pos_tag']

# Iterate through each document and pull out relevant tags
for ix, doc in enumerate(fake_spacydocs):
    tags = extract_token_tags(doc) # gets tags for each token
    tags = pd.DataFrame(tags) # puts info from above in table structure
    tags.columns = columns # adds column names to data table
    fake_tagsdf.append(tags) 

# fake_tagsdf is still in a list so we need to get in the right format and assign it back to fake_tagsdf
fake_tagsdf = pd.concat(fake_tagsdf)

fact_tagsdf = []
# Iterate through each document and pull out relevant tags
for ix, doc in enumerate(fact_spacydocs):
    tags = extract_token_tags(doc)
    tags = pd.DataFrame(tags)
    tags.columns = columns
    fact_tagsdf.append(tags) 

# gets it back to right function and assign it back to fake_tagsdf
fact_tagsdf = pd.concat(fact_tagsdf)
```

```python
# Removing everything before first hyphen
data['text_clean'] = data.apply(lambda x: re.sub(r"^[^-]*-\s", "", x['text']), axis=1)
data.head()
```

```python
# Making everything lowercase
data['text_clean'] = data['text_clean'].str.lower()
data.head()
```

```python
# Removing punctuation
data['text_clean'] = data.apply(lambda x: re.sub(r"([^\w\s])", "", x['text_clean']), axis=1)
data.head()
```

```python
# Removing stop words:
en_stopwords = stopwords.words('english')
print(en_stopwords)
```

```python
data['text_clean'] = data['text_clean'].apply(lambda x: ' '.join(word for word in x.split() if word not in en_stopwords))
data.head()
```

```python
# Tokenize text - go through each row and convert to word token
data['text_clean'] = data.apply(lambda x: word_tokenize(x['text_clean']), axis=1)
```

```python
# Lemmetize
lemmetizer = WordNetLemmatizer()
data['text_clean'] = data['text_clean'].apply(lambda tokens: [lemmetizer.lemmatize(token) for token in tokens]) 
data.head()
```

```python
#vectorize text:
fake_news_text = data[data['fake_or_factual'] == "Fake News"]['text_clean'].reset_index(drop=True)

# fake_news_text - list of clean preprocessed text list:
# 0     [two, small, problem, analogy, susan, jesus, m...
# 1     [buried, trump, bonkers, interview, new, york,...
# 2     [woman, make, 50, percent, country, grossly, u...

dictionary_fake = corpora.Dictionary(fake_news_text)
# dictionary_fake:
# Dictionary<6418 unique tokens: ['16yearold', '69yearold', 'accompanying', 'actor', 'actress']...>

#bag of words
doc_term_fake = [dictionary_fake.doc2bow(text) for text in fake_news_text]
# doc_term_fake - The occurrences of each word in a document are counted, order not considered:
# [[(0, 1), (1, 1), (2, 1), (3, 1), (4, 4), (5, 1), (6, 1), (7, 1), (etc)]] - 4th word occurs 4 times
```

```python
coherence_values = []
model_list = []

min_topics = 2
max_topics = 11

for num_topics_i in range(min_topics, max_topics+1):
    model = gensim.models.LdaModel(doc_term_fake,
                                   num_topics=num_topics_i,
                                   id2word = dictionary_fake) #lda model
    model_list.append(model)
    coherence_model = CoherenceModel(model=model,
                                    texts=fake_news_text,
                                    dictionary=dictionary_fake,
                                    coherence='c_v')
    coherence_values.append(coherence_model.get_coherence())
```

```python
plt.plot(range(min_topics, max_topics+1), coherence_values)
plt.xlabel('Number of Topics')
plt.ylabel('Coherence Scores')
plt.legend(('cohernece_values'), loc='best')
plt.show()
```

```python
num_topics_lda = 4
lda_model = gensim.models.LdaModel(corpus=doc_term_fake,
                                   id2word=dictionary_fake,
                                   num_topics=num_topics_lda)

# print 10 most common words that occur with each topic
lda_model.print_topics(num_words=10)
```

```python
# function to create tfidf corpus:
def tfidf_corpus(doc_term_matrix):
    tfidf = TfidfModel(corpus=doc_term_matrix,
                       normalize=True)
    corpus_tfidf = tfidf[doc_term_matrix]
    return corpus_tfidf
```

```python
# function to get coherence score
def get_coherence_score(corpus, dictionary, text, min_topics, max_topics):
    coherence_values = []
    model_list = []
    for num_topics_i in range(min_topics, max_topics+1):
        model = LsiModel(corpus, num_topics=num_topics_i, id2word=dictionary)
        model_list.append(model)
        coherence_model = CoherenceModel(model=model, texts=text, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherence_model.get_coherence())

    plt.plot(range(min_topics, max_topics+1), coherence_values)
    plt.xlabel("Number of Topics")
    plt.ylabel("Coherence Score")
    plt.legend(('coherence_values'), loc='best')
    plt.show()
```

```python
corpus_tfidf_fake = tfidf_corpus(doc_term_fake)
```

```python
get_coherence_score(corpus_tfidf_fake, dictionary_fake, fake_news_text, min_topics=2, max_topics=11)
```

```python
lsa_model= LsiModel(corpus_tfidf_fake, id2word=dictionary_fake, num_topics=3)
lsa_model.print_topics()
```

```python
# text_clean column will become x, which we will then vectorize
x = [','.join(map(str,l)) for l in data['text_clean']]
y = data['fake_or_factual']
```

```python
# using bag of words to vectorize text
countvec = CountVectorizer()
countvec_fit = countvec.fit_transform(x)

bag_of_words = pd.DataFrame(countvec_fit.toarray(), columns=countvec.get_feature_names_out())
bag_of_words
```

```python
# 30% of data will go to test set and will not be used for training
x_train, x_test, y_train, y_test = train_test_split(bag_of_words, y, test_size=0.3)
```

```python
# Using Logistic Regression:
lr = LogisticRegression(random_state=0).fit(x_train, y_train)
y_pred_lr = lr.predict(x_test)
accuracy_score(y_pred_lr, y_test)
```

```python
print(classification_report(y_test, y_pred_lr))
```

```python
# Using SVM
svm = SGDClassifier().fit(x_train, y_train)
y_pred_svm = svm.predict(x_test)
accuracy_score(y_pred_svm, y_test)
```

```python
print(classification_report(y_test, y_pred_svm))
```
